## Duplicates

```python
def duplicates(df):
    return pd.DataFrame(df[~(df.duplicated() | 
                             df.duplicated(subset=['SNPID_UKB']) | 
                             df['SNPID_UKB'].isnull() |
                             df.duplicated(subset=['SNP']) |
                             df['SNP'].isnull())].reset_index(drop=True))

# there are no duplicates
print("Duplicates:")
print("Systolic:", sdf.shape[0] - duplicates(sdf).shape[0])
print("Diastolic:", ddf.shape[0] - duplicates(ddf).shape[0])
print("Pulse Rate:", pdf.shape[0] - duplicates(pdf).shape[0])
```

## Creating S_XY_FULL_STUDY_1 for S_YY

```python
if not os.path.isdir(os.getcwd()+"/data/processed"):
    os.mkdir(os.getcwd()+"/data/processed")
```

While creating s_xy_full_study_1, A1 column is the effect allele/ALT allele and A2 is the non-effect allele/REF allele. <br>
So, the format for allele_0 and allele_1 would be similar to vcf format of REF first and ALT second.<br>
A2 and A1 are positioned in the dataframe accordingly.

```python
sxy_full_study_1 = pd.DataFrame(sdf[['SNP','SNPID_UKB','A2','A1','BETA','SE']].rename(columns={'A2':'allele_0',
                                                                                               'A1':'allele_1',
                                                                                               'BETA':'sys_b',
                                                                                               'SE':'sys_se'}))
sxy_full_study_1
```

```python
sxy_full_study_1 = sxy_full_study_1.merge(ddf[['SNP', 'BETA', 'SE']].rename(columns={'BETA':'dia_b',
                                                                                     'SE':'dia_se'}), 
                on='SNP', how='inner').reset_index(drop=True)

sxy_full_study_1.head()
```

```python
sxy_full_study_1 = sxy_full_study_1.merge(pdf[['SNP', 'BETA', 'SE']].rename(columns={'BETA':'pul_b',
                                                                                     'SE':'pul_se'}), 
                on='SNP', how='inner').reset_index(drop=True)

sxy_full_study_1.head()
```

```python
sxy_full_study_1.drop(columns=['SNP'], inplace=True)
sxy_full_study_1.reset_index(drop=True, inplace=True)
```

```python
sxy_full_study_1.info()
sxy_full_study_1.head()
```

```python
sxy_full_study_1.to_csv("./data/processed/sxy_full_study_1.txt", sep="\t", index=False)
```

## Creating CHR,BP,SNP,ALLELEs VCF file 

```python
# getting rsIDs for SNP and saving files for vcf format
temp_df = sdf[sdf['MAF'] > 0.01]
temp_df = pd.DataFrame(
    {
        '#CHROM':temp_df['CHR'],
        'POS':temp_df['BP'].astype(int),
        'ID':temp_df['SNPID_UKB'],
        'REF':temp_df['A2'],
        'ALT':temp_df['A1'],
    }
)

temp_df['QUAL'] = ['NA']*temp_df.shape[0]
temp_df['FILTER'] = temp_df['QUAL']
temp_df['INFO'] = temp_df['QUAL']

print("Headers (inorder):", temp_df.columns.values.tolist())
temp_df.to_csv(os.getcwd()+'/data/processed/full_data.vcf', sep='\t', index=False, header=False)

del temp_df
```
